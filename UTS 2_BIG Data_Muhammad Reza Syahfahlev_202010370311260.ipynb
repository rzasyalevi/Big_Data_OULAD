```python
bd = bd.dropna(subset=['Destination'])
```

```python
bd.isnull().sum()
```

```python
def encoding(list_encode):
  label_encoder = LabelEncoder()
  for i in list_encode:
    for j in bd:
      if i == j:
        bd[i] = label_encoder.fit_transform(bd[i])
  return bd
```

```python
list_encode = ['Destination', 'Protocol', 'Length', 'Info', 'Source']
```

```python
encoding(list_encode)
```

```python
correlation_matrix = bd.corr()
relevant_features = correlation_matrix['Protocol'].sort_values(ascending=False)
relevant_features
```

```python
# Split features and labels
X = bd[['Length', 'Destination', 'Time', 'No.']].values
y = bd['Protocol'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

# Naives Bayes


```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import datasets

model = GaussianNB()

# Latih model menggunakan data latih
model.fit(X_train, y_train)

# Prediksi kelas pada data uji
y_pred = model.predict(X_test)

# Evaluasi model (misalnya, akurasi)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f'Akurasi: {accuracy:.2f}')
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
# Inisialisasi model Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Latih model menggunakan data latih
model.fit(X_train, y_train)

# Prediksi kelas pada data uji
y_pred = model.predict(X_test)

# Evaluasi model (misalnya, akurasi)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f'Akurasi: {accuracy:.2f}')
```
