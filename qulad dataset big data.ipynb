Assessments

```python
#membaca dataset assessments
ed = pd.read_csv('assessments.csv')
ed
```

```python
# menampilkan semua informasi data

ed.info()
```

```python
#membaca minmaxcountmean dan lain lain

ed.describe()
```

```python
#membaca summary total data null
print(ed.isnull().sum())
```

```python
#Dilakukannya pengisian data yang null dengan nilai rata rata/ nilai yang sering muncul
for col in ed.columns:
    mode_value = ed[col].mode()[0]
    ed[col].fillna(mode_value, inplace=True)
print(ed.isnull().sum())
```
